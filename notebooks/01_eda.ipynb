# Exploratory Data Analysis (EDA)

This notebook demonstrates how to load the processed data and perform basic analysis.

## Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load processed data
df = pd.read_csv('../data/processed/inventory_data_processed.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

## Basic Stats

```python
df.describe()
```

## Sales Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title('Distribution of Daily Sales')
plt.show()
```
